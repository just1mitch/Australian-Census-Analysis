## **Generic Functions**



```python
import matplotlib.pyplot as plt
from matplotlib import colors as mcolours
import matplotlib.image as mpimg
import numpy as np
```

#### **Colour Mapping**



```python
# Imports matplotlib color profile and creates a dictionary that
# maintains the colours of categories across datasets
#
# Means that a category has one consistent colour across the whole notebook
colour_dict = {}
for colour in mcolours.TABLEAU_COLORS: # TABLEAU only supports 10 colours - enough for the analysis
    colour_dict[colour] = None
```

#### **Removing Top 'Junk' Rows and Creating Headers**



```python
# Takes a list of csv lines, removes the top n 'junk' lines, and isolates the headers
# Takes 2 arguments:
#     - data_list is the list of csv line 'raw data'
#     - n is the number of initial rows to remove from the raw data
# Returns 2 variables:
#     - headers is a list of categories in the dataset
#     - data_list is the list of rows containing actual data
def get_headers(data_list, n):
    data_list = data_list[n:]
    headers = []
    header_rows = 0 # Number of rows taken up by headers, based on ABS format
    for string in data_list:
        if (string[0] != ","):
            data_list = data_list[header_rows:]
            for index, header in enumerate(headers):
                headers[index] = header[:-1] # Remove the final extra " "
            return headers, data_list
        else:
            header_rows += 1
            string_list = string.split(",")
            string_list = string_list[1:] # Remove first element as this is never a header
            while (len(headers) != len(string_list)):
                headers.append("")
            for index, item in enumerate(string_list):
                if len(item) > 0:
                    headers[index] = headers[index] + item + " "
    return headers, data_list # Only reach here if data_list is empty
```

#### **Convert Data Strings into Floats**



```python
#Take input as list of all data.readlines() strings
#cleanes the numerical data, from strings to float
#Return a dictionary containing key- price range and values-numerical data which is sorted in the order of the header it corresponds to
def make_float(inp):
    var = [] #independent variables (price range of rent)
    cleaned = [] #clean each line
    final = [] #from cleaned list, convert to correct type
    for l in inp:
        var.append(l.split(',')[0]) #even the numbers with 
        clean = change_line(l.strip()) #if any double quotes contained, send to function to clean
        cleaned.append(clean.split(','))
    for lists in cleaned: 
        floats = [float(num) for num in lists if num.strip() and num.isdigit()] #contain numerical data only
        if floats: #ensure non-empty, then append to final list
            final.append(floats)
    var = [item for item in var if item != ''][:len(final)] #take away excess description texts we've seen in data 
    d = {} #producing a dictionary will make it easier to create visual interpretation as the keys and values will correspond to one another
    for v in var:
        d[v] = final[var.index(v)] #allocate each variable the list of its numerical data
    return(d)

#Input is a string of a singular line of data from file that has been stripped
#Ths function calls itself until required cleaning is complete
#Output is a string of the input with the double quotes removed and any commas between numbers >999 removed
def change_line(inp): #make a generic function to accomodate the discrepancies in the data
    if '"' in inp: #focusing on the layout of the data, we need to get rid of, numerical values in double quotes
        starting = inp.index('"')
        ending = inp.index('"', starting + 1)
        n = inp[starting + 1:ending].strip('"').replace(',', '') 
#the leading and trailing double quotes and the comma in between numbers of value thousand and above are removed
        fixed = inp[:starting] + n + inp[ending + 1:]
        return change_line(fixed) #continue until we can split string commas without disrupting the data
    return inp
```

#### **Open Datafile and Parse Data**



```python
# Gets lines of data from a file
# Takes 2 arguments:
#     - DATAFILE is the file to read data from
#     - junk_lines is the number of initial rows to remove from the data
# Returns 4 variables:
#     - headers is a list of categories in the data
#     - value_dict is a dictionary of the data, with keys as the weekly cost bracket and values as the count for each category
#     - totals is a list of tuples of the format (category - (Distribution %), Total count for category across all cost brackets)
#     - per is a list of tuples in the format (Distribution (as a decimal), category)
def get_data(DATAFILE, junk_lines):
    with open(DATAFILE) as data:
        line = data.readlines()
        headers, lines = get_headers(line, junk_lines)
        value_dict = make_float(lines)
        headers = [item.strip().upper() for item in headers]
        list_total = value_dict['Total']
        total_renters = list_total[-1]
        per = [(round(num/total_renters, 4), headers[list_total.index(num)]) for num in list_total[:-1]]
        # totals is a list of lists in the format (rental type - (% of total), total of rental type)
        # colour of line is stored in global dictionary for continuity across graphs
        totals = [[f"{header} - ({round(value_dict['Total'][i] * 100 / total_renters, 1)}%)", value_dict['Total'][i]] for i, header in enumerate(headers)][:-1]
        try:
            headers.remove("TOTAL")
        except ValueError:
            pass
        del value_dict['Total']
    return headers, value_dict, totals, per
```

#### **Change Header to New Header**



```python
# Renames a header in the dataset
# Takes 5 variables:
#     - old_header is the original name of the header in the data
#     - new_header is the name that old_header is renamed to
#     - headers is a list, unchanged except for the specified header name
#     - totals is the list of tuples, also unchanged except for the header name
#     - percentages is another list of tuples, also unchanged except for the header name
# Returns 3 variables:
#     - headers, the same variable passed to the function except with the changed header name
#     - totals, the same variable passed to the function except with the changed header name
#     - percentages, the same variable passed to the function except with the changed header name
def change_header(old_header, new_header, headers, totals, percentages):
    # Size and order of headers, totals and percentages is assumed to be identical
    for i in range(len(headers)):
        # Change in headers list
        if headers[i] == old_header:
            headers[i] = new_header
            percentages[i] = (percentages[i][0], new_header)
            totals[i][0] = new_header + f" - ({round(percentages[i][0] * 100, 1)}%)"
            return headers, totals, percentages
    return headers, totals, percentages # no data changed
```

#### **Create Line Graph of Weekly Rental Prices**



```python
# Creates a line graph for the given data
# Takes 3 arguments:
#     - headers, a list of category names
#     - value_dict is a dictionary of the data, with keys as the weekly cost bracket and values as the count for each category
#     - YEAR, the ABS census year the data comes from
def create_line_graph(headers, value_dict, YEAR):
    for index, category in enumerate(headers):
        y = [(list[index]) for list in value_dict.values()] # Relies on dictionary entries being in order, which is Python standard as of 3.7
        # Makes sure category has an assigned colour
        colour_assigned = False
        for entry in colour_dict.items():
            if (entry[1] == category):
                chosen_colour = entry[0]
                colour_assigned = True
                # Plot the line for this category
                plt.plot(value_dict.keys(), y, 'o', ls = '-', label = f"{category}", color = chosen_colour)
        if not colour_assigned:
            # Assign a colour from colour_dict global dictionary
            for colour in colour_dict.keys():
                if colour_dict.get(colour) == None:
                    colour_dict[colour] = f"{category}"
                    colour_assigned = True
                    chosen_colour = colour
                    break
            if not colour_assigned:
                print("Error: Ran out of assignable colours")
                return
            else:
                plt.plot(value_dict.keys(), y, 'o', ls = '-', label = f"{category}", color = chosen_colour)

    # Formatting of the graph
    plt.title(f"{YEAR} - Weekly Rental Cost Distribution by Landlord Type", color = "Black", size = 20)
    plt.xlabel("Weekly Rent ($)", color = "Black", size = 12)
    plt.ylabel("Number of Renters", color = "Black", size = 12)
    plt.tick_params(colors = "Black")
    plt.xticks(rotation=45)
    plt.legend(title = "Landlord Type", title_fontproperties = {'weight': 'bold'})
    plt.show()
    return
```

#### **Create a Pie Chart from Totals**



```python
# Removes the distribution percentage from a header title
# Takes 1 argument:
#     - header, in the format (category - (DISTRIBUTION %))
# Returns 1 argument:
#     - header, no longer containing the distribution %
# Used solely by create_pie function below
def remove_percentage(header):
    # Remove anything after the final '-', which indicates the percentage
    header_backwards = header[::-1]
    for index, letter in enumerate(header_backwards):
        if letter == '-':
            # grab all letters following before the whitespace before the -
            new_header = header[:len(header)-index-2]
            return(new_header)
    return header
        
# Creates a pie chart showing the distribution of landlord type
# Takes 2 arguments:
#     - totals is a list of tuples of the format (category - (Distribution %), Total count for category across all cost brackets)
#     - YEAR, the ABS census year the data comes from
def create_pie(totals, YEAR):
    totals.sort(reverse = True, key = lambda x:x[1])
    pie_values = []
    pie_categories = []
    pie_colours = []
    for list in totals:
        pie_categories.append(list[0])
        pie_values.append(list[1])
        colour_found = False
        header = remove_percentage(list[0])
        for assigned_colours in colour_dict.items():
            if assigned_colours[1] == header:
                pie_colours.append(assigned_colours[0])
                colour_found = True
                break
        if not colour_found:
            print(f"Error: Category {header} not assigned a colour")
            return
    
    plt.pie(x = pie_values, labels = pie_categories, colors = pie_colours, textprops = {'color': 'Black'})
    plt.title(f"{YEAR} - Distribution of Renters across Landlord Types", color = "Black", size = 20)
    plt.savefig(f"./graphs/{YEAR} pie.png", bbox_inches = 'tight')
    plt.show()
```

#### **Trend 2: Sharp Decrease in** _**Other**_ **Category After 2001**

705,122 of 1,858,324 renters in 2001 identified under the _other_ category \- making up 37.9% of the population. However, this proportion dropped drastically to 99,434 out of 2,010,456 renters \(only ~5%\). This is a direct result of a reclassification of the category after 2001. In 2001, _Other_ included all rental properties that were not managed by a private landlord or a state/territory housing authority. Realistically, this initial structure failed to capture the demographic of rental properties in Australia at the time, as there were clearly too many housing situations that did not fit into the binary categories of 'public or private' rental properties. The ABS subsequently added 2 new categories for the 2006 census and beyond \- _Person not in same household_ and _community housing_. 

The new definition of _Other_ became "dwellings being rented through a 'residential park \(includes caravan parks and marinas\)', private employer or government employer \(includes Defence Housing Authority\)". The reclassification provided better insight into the rental distribution of Australia, and since 2006 the _Other_ category has seen a consistent proportion of between 4\-5% of all rental properties. 

The line graph below highlights the impact of the introduction of the new categories in 2006. The increase in proportion of the _real estate agent_ category, mentioned in **Trend 1**, is also visualized.

In 2001 and 2021, rent\-free dwellings are explicitly excluded from the data collection. However, in 2006, 2011 and 2016, households where tenants are living rent free are included. You can verify this by viewing the first datapoint of each year's line graph \- the first weekly rent bracket for 2006 is from '0\-49', 2011 & 2016 start from '0\-74' whilst the first 2001 rent range is '1\-49' and 2021 starts from '1\-74' \(all in \$AUD\).

Whilst data for _Person not in same household_ was not recorded in 2001, we can observe the impact this has on the change in proportion from 2006/2011/2016 and 2021.

2006 \(Rent\-Free **Included**\): 23.6% of all rental properties are _Person not in same household_

2011 \(Rent\-Free **Included**\): 22.7% of all rental properties are _Person not in same household_ \(down <span style='color:#f44336'>0.9%</span>\)

2016 \(Rent\-Free **Included**\): 20.2% of all rental properties are _Person not in same household_ \(down <span style='color:#f44336'>2.5%</span>\)

2021 \(Rent\-Free **Excluded**\): 17.4% of all rental properties are _Person not in same household_ \(down <span style='color:#f44336'>2.8%</span>\)

Whilst we can see a 2.8% decrease in proportion after excluding rent\-free households, we see a similar decrease in proportion from 2011 to 2016. It is more insightful to consider the proportion of households categorized as _Person not in same household_ when only considering households in the _first data bracket_, without factoring in more expensive rental properties:

2006 \(Rent\-Free **Included**\): 30733 of 111675 \(27.5%\) rental properties in \$0\-49 range are _Person not in same household_

2011 \(Rent\-Free **Included**\): 39444 of 155466 \(25.4%\) rental properties in \$0\-74 range are _Person not in same household_ \(down<span style='color:#f44336'> 2.1%</span>\)

2016 \(Rent\-Free **Included**\): 39136 of 116613 \(33.6%\) rental properties in \$0\-74 range are _Person not in same household_ \(up <span style='color:#4caf50'>8.2%</span>\)

2021 \(Rent\-Free **Excluded**\): 3806 of 29839 \(12.8%\) rental properties in \$0\-74 range are _Person not in same household_ \(down <span style='color:#f44336'>20.8%</span>\)

The total number of households in the category and cost bracket decreases by almost a factor of 10 from 2016 to 2021 \(some of this decrease is also a result of inflation increasing weekly rent cost, moving some houses to higher cost brackets\). Also note that despite the overall proportion of the category decreasing across the cumulative cost brackets from 2011 to 2016 by 2.5%, in the \$0\-\$74 bracket the proportion actually grew by 8.2%.

The inclusion of rent\-free dwellings can skew the data for each landlord type \- as these households will typically fall under the _Person not in same household_ landlord type. Public housing \(owned by state/territory authorities\) are typically cheaper on a weekly basis than private housing \(through Real Estate Agents\), but are not \$0 a week. Households in the _Person not in same household_ category are those rented from friends or family. Contextually, it makes logical sense that some people may be living under a relative or friend's house without paying a weekly rental fee.

Without making a comment on whether or not rent\-free households _should_ be included in the data, it is clear how impactful the decision is on the interpretations that can be made from the data. 



#### **Sources of Error**

As previously addressed, a major error is the categories and their definition. In 2001, the 'Other' category included 'Housing Co\-operative Community/Church Group' and 'Person Not in Same Household' however from 2006 and onwards, there are seperate categories for them. Such transition in categorization introduces a challenge in data analysis to accommodate it causing additional scrutiny during data processing. This would require extra observation and when compiling an analysis, can easily be overlooked on accident. A form of improving this analysis of this error would be cross checking with the source. ABS has given a comprehensive [list](https://www.abs.gov.au/census/guide-census-data/census-dictionary/2021/variables-topic/housing/landlord-type-lldd) of landlord types as well as previous changes made. Reading the variables from the data and ensuring the data for each year is categorised correctly allows optimal results.

Within the trends, it was noted how the definition of what is considered a rental across years eventually changed. From the cateogry _Person Not in Same Household_, the minimum threshold range increasing from 0 to 1 excluded the community of rent\-free dwellers. This is simply one observation made through analysing the visual data. It is possible there are more discrepancies in the definition of other categories throughout the year. If the analysis fails to pick up and account for any changes, the interpretation can be considered less valid as major factors are left out. 

Another important error regards in general, the inclusion of data. It should be noted that the datapack for 1996 did not include weekly rental cost by landlord type. This is counted as an error within the analysis because there could show lack of awareness into important trends or patterns that could have been established during that period for the rest of the data. In an ideal scenario, all sets having valid datapacks would be highly preferred to ensure a comprehensive analysis as the unavailability of data for one of the years in this analysis posed a limitation and restricted the scope of the analysis. When looking at the question this analysis is prompted to answer, the change of distribution is asked over a span of 20 years. This is fitting as the data covers the timeline, however a span of 25 years would be ideal to give a deeper insight.
